## This python script is for the following purposes:
- To remove encryption from the pdf
- To remove the layering in the pdf
- To remove the digital signatures in the file
- To remove security headers in the pdf

### Result:
- A new pdf is generated which can be used to upload on website, which do not accept the pdf's with digital security headers or multiple layered pdf's.
- Secure as it is offline and no data will be uploaded anywhere. Best method if you prefer privacy.
Some document examples - Bank Statements, Payslips, Agreements etc.

```python
# Run this command to install PyPDF2 if you haven't already:
pip install PyPDF2
```

```python
from PyPDF2 import PdfReader, PdfWriter
```

```python
input_path = "Document_name.pdf"
output_path = "Document_Cleaned.pdf"
```

```python
# Open the original PDF in read-binary mode
with open(input_path, "rb") as infile:
    reader = PdfReader(infile)

    # If the PDF is actually password-protected, you'd need:
    # reader.decrypt("your_password_here")

    writer = PdfWriter()

    # Copy each page into a new writer
    for page in reader.pages:
        writer.add_page(page)

    # Write everything out to a brand new PDF
    with open(output_path, "wb") as outfile:
        writer.write(outfile)
    print("Done! Clean file saved as:", output_path)
```
